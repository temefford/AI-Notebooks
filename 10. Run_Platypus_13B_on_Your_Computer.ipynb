In this notebook, we use 4-bit quantization to run Platypus2-13B. This code uses only 12 GB of VRAM. It can run on a free instance of Google Colab or on a local GPU (e.g., RTX 3060 12GB).
More details here: [Platypus: Dataset Curation and Adapters for Better Large Language Models](https://kaitchup.substack.com/p/platypus-dataset-curation-and-adapters)

We only need the following libraries:


*   transformers
*   accelerate (for device_map)
*   bitsandbytes (for 4-bit quantization)

With device_map="auto" and load_in_4bit, the model consumes only 11 GB of VRAM.
Note that Platypus is specialized in the STEM domain, so if your prompt is out of the STEM domain, you may get very bad answers.


```python
from transformers import AutoModelForCausalLM, AutoTokenizer
model_name = "garage-bAInd/Platypus2-13B"
prompt = "Tell me about gravity"

model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", load_in_4bit=True)
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
model_inputs = tokenizer(prompt, return_tensors="pt").to("cuda:0")

output = model.generate(**model_inputs)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```
